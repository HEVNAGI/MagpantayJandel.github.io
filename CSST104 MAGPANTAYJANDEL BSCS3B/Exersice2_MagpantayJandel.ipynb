**1. Creating Dataset**

```python
import pandas as pd

data = {
    'student_id': range(1,16),
    'gender': ['Male','FeMale','Male','FeMale','FeMale','Male','FeMale','Male','FeMale','Male','FeMale','Male','FeMale','Male','FeMale'],
    'test_date': pd.date_range(start = '2024-02-14', periods=15),
    'study_hours': [5,3,4,5,6,7,8,9,1,2,3,5,6,4,7],
    'test_score': [80,75,76,78,89,90,98,74,84,81,82,83,91,92,93],
    'attendance': ['Present','Absent','Present','Present','Absent','Present','Absent','Present','Absent','Absent','Present','Absent','Present','Absent','Present'],
}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

**2. Data Loading and Initial Examination**

```python
df = pd.read_csv('student_performance.csv')

print(df.head(10))

print(df.isnull().sum())
```

**3. Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score','study_hours']].describe()

#Histogram
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sne.histplot(df['test_score'],bins=15,kde=True)
plt.title('Distribution of Test Score')

plt.subplot(1,2,2)
sne.scatterplot(data=df,x='study_hours',y='test_score')
plt.title('Distribution of study hours')

#scatterplot
plt.figure(figsize=(10,8))
sne.scatterplot(data=df ,x='study_hours' ,y='test_score')
plt.title('Scatter Plot of Study Hours vs. Test Scores')
```

**4. Data Visualization**

```python
plt.figure(figsize=(6,4))
sne.countplot(data=df ,x= 'gender')
plt.title('Gender Distribution')

plt.figure(figsize=(10,6))
sne.lineplot(data=df ,x='test_date' ,y='test_score')
plt.title('Trend in Test Score Over time')

correlation_matrix =df[['study_hours','test_score','attendance']].corr()
sne.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```
