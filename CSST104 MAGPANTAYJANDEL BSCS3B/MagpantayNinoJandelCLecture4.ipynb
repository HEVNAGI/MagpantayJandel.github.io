1. Import Libraries

```python
import pandas as pd #Data Manipulation
import numpy as np #Mathematical operation
import matplotlib.pyplot as plt #Data Vizualization
from sklearn.linear_model import LinearRegression #Training Data/Data Modeling
```

```python
df = pd.read_csv('Jandel.csv')
print(df)
```

```python
plt.scatter(df['Previous_Exam_Scores'],df['Grades'])
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Previous Exam Scores')
plt.show()
```

```python
x = df[['Previous_Exam_Scores']]
y = df[['Grades']]
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Intercept:", slope)
print("Intercept:", intercept)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='pink')
plt.xlabel('Independent Variable(x)')
plt.ylabel('Dependent Variable(y)')
plt.title('Lecture 4')
plt.show()
```

```python
new_x = [[np.random.randint(50,100)]]
predicted_y = model.predict(new_x)
print("value of x:", new_x)
print("Predicted y for new x:", predicted_y)
```
